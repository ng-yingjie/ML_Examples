## K-Means Clustering 

```python
import pandas as pd
import numpy as np
import time

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def kmeans_measurement(data, n):
    kmeans = KMeans(n_clusters= n)
    kmeans.fit(data[0])
    kmeans.cluster_centers_
    return (kmeans.labels_ )
```

```python
def compare_plots(data, kmeans_label):
    fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
    ax1.set_title('K Means')
    ax1.scatter(data[0][:,0],data[0][:,1],c = kmeans_label,cmap='rainbow')
    ax2.set_title("Original")
    ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
    return 
```

### Changing cluster std: Change distance between each cluster points

```python
std_1 = make_blobs(n_samples=200, n_features=2, centers= 5, cluster_std=10)
std_2 = make_blobs(n_samples=200, n_features=2, centers= 5, cluster_std=1)
```

```python
n = 5
results_std_1 = kmeans_measurement(std_1, n)
results_std_2 = kmeans_measurement(std_2,n)
```

```python
compare_plots(std_1, results_std_1)
```

```python
compare_plots(std_2, results_std_2)
```

**K-Means seem to work better for datapoints with smaller variances within samples (std smaller)** 

Large variance = many points being scattered and less able to calculate true nearest neighbour within cluster 

### Changing cluster size

```python
cluster_1 = make_blobs(n_samples=500, n_features=2, centers= 20, cluster_std=1.5)
cluster_2 = make_blobs(n_samples=500, n_features=2, centers= 5, cluster_std=1.5)
```

```python
results_cluster_1 = kmeans_measurement(cluster_1, n =20)
results_cluster_2 = kmeans_measurement(cluster_2, n =5)
```

```python
# 25 clusters
compare_plots(cluster_1,results_cluster_1)
```

```python
# 5 clusters
compare_plots(cluster_2,results_cluster_2)
```

**Similar to large std, more clusters results in overlaps of points and therefore lower ability to differentiate between clusters**

### Changing Number of Features

```python
feature_1 = make_blobs(n_samples=200, n_features=10, centers= 5, cluster_std=1.5)
feature_2 = make_blobs(n_samples=200, n_features=2, centers= 5, cluster_std=1.5)
```

```python
# Let's explore what n_features really mean
feature_1
```

```python
feature_2
```

Seems like features = number of columns or can be understood as the number of variables 

```python
n = 5
results_feature_1 = kmeans_measurement(feature_2, n)
results_feature_2 = kmeans_measurement(feature_2, n)
```

```python
compare_plots(feature_1,results_feature_1)
```

```python
compare_plots(feature_2,results_feature_2)
```

**Increase of the number of features increases the dimensionality used to determine the clusters. It may have complicated calculation of Euclidean distance and therefore giving overlaps of clusters**

k-Means work best for 2d/3d dataframes (https://towardsdatascience.com/k-means-clustering-implementation-2018-ac5cd1e51d0a)

### Changing Sample Size

```python
time_start = time.clock()
size_1 = make_blobs(n_samples=9999, n_features=3, centers= 5, cluster_std=1.5)
results_size_1 = kmeans_measurement(size_1, n =5)
compare_plots(size_1, results_size_1)
time_elapsed = (time.clock() - time_start)
```

```python
time_start = time.clock()
size_2 = make_blobs(n_samples=200, n_features=3, centers= 5, cluster_std=1.5)
results_size_2 = kmeans_measurement(size_2, n =5)
compare_plots(size_2, results_size_2)
time_elapsed = (time.clock() - time_start)
```

**Size of datasets don't seem to matter here except to increase computational time to 'create' and reduce clusters and measure observation to centroid distance**
